#### Generate randomly monthly electricity bill of a year then plot the bill using using line, bar and pie chart

```python
import matplotlib.pyplot as plt
import numpy as np 
from numpy import random
randomBill = random.randint(1000,3000,size=(12))
month = np.array(['Jan','Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.plot(month,randomBill)
plt.show()
```

```python
plt.pie(randomBill, labels=month)
plt.show()
```

```python
plt.bar(month,randomBill)
plt.show()
```

#### Generate 30 random 2-d points and draw it using scatter plot

```python
xpoints = random.randint(100,size=(30))
ypoints = random.randint(100,size=(30))
plt.scatter(xpoints, ypoints)
plt.show()
```

#### Draw Sin and Cos curve

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.show()
```
